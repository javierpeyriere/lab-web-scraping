````python
def extract_allo_cine_info(soup):
    # your code goes here
    
    # Lists to store extracted data
    movie_numbers = []
    titles = []
    years = []
    ratings = []
    runtimes =[]
    genres = []
    descriptions = []
    directors = []
    stars = []
    votes = []
    #gross_earnings = []

    # Find all components
    components = soup.find_all('li', class_='mdl')

    for component in components:
        # Extract movie number:
        movie_number = component.find('a','meta-title-link').get('href')
        temp = movie_number.replace('/film/fichefilm_gen_cfilm=','').replace('.html','')
        movie_numbers.append(temp)
        
        # Extract movie name
        title = component.find('span').get('title')
        if title.startswith('Bande-annonce '):
            title = title.replace('Bande-annonce ', '')
        titles.append(title)
        
        # Extract date      
        year = component.find('span', class_='date').text.strip()
        # Converting the date:
        months_french = ['janvier', 'février', 'mars', 'avril', 'mai', 'juin', 'juillet', 'août', 'septembre', 'octobre', 'novembre', 'décembre']
        temp = [int(year[:2]),"",int(year[-4:])]
        for month in months_french:
            i=0
            if month == months_french[i]:
                temp[1] = i+1
            i+=1
        d, m, y = temp
        years.append(datetime(y, m , d))
        
        # Extract rating     
        rating = component.find('span', class_='stareval-note').text.strip()
        ratings.append(float(rating.replace(",",".")))

        # Extract runtime, genre     
        runtime = component.find('div', class_='meta-body-item meta-body-info').text.strip().replace(',','')
        temp = runtime.split('\n')
        runtimes.append(temp[2])
        genres.append(temp[4:])

        # Extract description     
        description = component.find('div', class_='content-txt').text.strip()
        descriptions.append(description)

        # Extract director     
        director = component.find('div', class_='meta-body-item meta-body-direction').text.strip().replace(',','')
        temp = director.split('\n')
        directors.append(temp[1:])

        # Extract star     
        star = component.find('div', class_='meta-body-item meta-body-actor').text.strip().replace(',','')
        temp = star.split('\n')
        stars.append(temp[1:])
        
        # Extract vote        
        vote = component.find('div', class_='rating-mdl').text.strip() #to put something
        #vote = compo.find('span', class_='stareval-review').text.strip() # tried but was not working
        votes.append(vote)    
             
    # Create DataFrame
    df = pd.DataFrame({
        'movie_numbers': movie_numbers,
        'titles': titles,
        'years': years,
        'rating': ratings,
        'tuntime': runtimes,
        'genres' : genres,
        'description' : descriptions,
        'directors' : directors,
        'stars' : stars,
        'votes' : votes,
        #'gross_($M)': gross_earnings,
        })
    
    # Filtering
    
    
    return df
````

````python
def souping(url):
    response = requests.get(url)
    response 
    soup = BeautifulSoup(response.content, "html.parser")
    return extract_allo_cine_info(soup)    
````

````python
# Your solution goes here

def next_page(url):
    # Finding the decade
    root = "https://www.allocine.fr/films/decennie-"
    temp = url.replace(root,"")
    decade = temp[:4]
    
    # Checking if it is the 1st page, and finding the page number to generate the next one
    endstring = temp.replace(decade,"").replace('/annee-','').split('/')
    
    if endstring[-1]== "":   # It is the first page
        new_url = url + "?page=2"
    else:
        to_add = "?page=" + str(int(endstring[-1].split('=')[-1])+1)
        new_url = url.split('?')[0] + to_add
    return new_url

def next_year(url):
    # Finding the year (on the first page)
    root = "https://www.allocine.fr/films/decennie-"
    temp = url.replace(root,"")
    year = temp[-5:-1]
    new_year = str(int(year)+1)+'/'
    temp= temp[:-5]
    new_url = root + temp + new_year
    return new_url

def next_decade(url):
    # Finding the decade
    root = "https://www.allocine.fr/films/decennie-"
    temp = url.replace(root,"")
    decade = temp[:4]
    new_decade = str(int(decade)+10)
    return url.replace(decade, new_decade)


    
````

````python
def the_data_for_one_year(url, user_rating = 0, n_page = 3, start_date = '1991-01-17' , end_date = '1992-10-17'):
    """Doing the function looking at only 3 pages for the moment, but apply the filter.
    Parameters: url -a year from allocine website-, user_rating = 0 float per default, n_page = 3 integer per default, 
    start_date = '1991-01-17' per default , end_date = '1992-10-17'per default)"""
    
    df = souping(url)
    mask = (df['years']>= start_date) & (df['years']< end_date) & (df['rating'] >= user_rating)
    df = df[mask]

    for i in range(0, n_page):
        url = next_page(url)
        df_page = souping(url)
        df_page = df_page[mask]
        df = pd.concat([df, df_page])
        print('step done ', i+1)
    return df.reset_index(drop = True)


````
